```python
# Load library
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Create URL to CSV file (alternatively this can be a filepath)
wbst = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'  
    
# Load the first sheet of the JSON file into a data frame
df = pd.read_csv(wbst)
# View the first ten rows
df.head(10)
```

```python
#find the columns by name 'unnamed' then drop those columns
df1 = df.drop(df.columns[df.columns.str.contains('unnamed',case = False)],axis = 1)
df1.head(10)
```

```python
with sns.axes_style('white'):
    g = sns.factorplot("Gender", data=df1, aspect=2,
                       kind="count", color='steelblue')
    g.set_xticklabels(step=5)
```

```python
df2 = (df1.groupby('Name')["Name"].count())
df2.nlargest(5)
```
